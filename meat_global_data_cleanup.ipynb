```python
kton = meat[meat["measure"]=="THND_TONNE"]
kgcap = meat[meat["measure"]=="KG_CAP"]
```

```python
all_countries = pd.read_csv("meat/world_countries-master/world_countries-master/data/countries/_combined/countries.csv")
code3 = all_countries[["alpha3","en"]]
```

```python
code3["alpha3"] = code3.alpha3.str.upper()
code3
```

```python
kton
```

```python
kton = kton.merge(code3,how="left", left_on='location', right_on='alpha3')
kton
```

```python
kton['en'].value_counts(dropna=False)
```

```python
kton = kton.drop(['indicator','measure','frequency'],axis=1 ) #remove all values same columns in kton
```

```python
kton.rename({'time':'Year','subject':'Species','en':"Country","value":"amount, kTn"}, axis=1, inplace=True) #renaming columns
```

```python
kton
```

```python
kton.Country.fillna(kton.location, inplace=True) 
kton
```

```python
kton.drop(['location','alpha3'],axis=1, inplace=True)
```

```python
# kton.rename({'Year':'year','Species':'species','Country':"country","amount, kTn":"amount"}, axis=1, inplace=True)
kton["amount"] = kton["amount"].round(2)
```

```python
kton = kton[(kton["country"]!='EU27') & (kton["country"]!='WLD') & (kton["country"]!='OECD') & (kton["country"]!='BRICS')]
```

```python
kton
```

```python
kton.to_csv("meat/kton_clean.csv")
```
